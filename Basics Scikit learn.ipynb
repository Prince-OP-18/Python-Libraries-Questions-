```python
import sklearn
from sklearn.datasets import load_iris  
X,y=load_iris(return_X_y=True)
```

```python
#Creation of Model-1
from sklearn.linear_model import LinearRegression
Model=LinearRegression()    
```

```python
#Fit data to Model-1
Model.fit(X,y)
#Check if model-1 runs
Model.predict(X)
```

```python
#Creation of Model-2
from sklearn.neighbors import KNeighborsRegressor
mod=KNeighborsRegressor()
#Fit the data to Model-2
mod.fit(X,y)
#Check if Model-2 runs
mod.predict(X)
```

```python
#Visualize Model-2
import matplotlib.pyplot as plt
pred=mod.predict(X)
plt.scatter(pred,y)
plt.show()
```
