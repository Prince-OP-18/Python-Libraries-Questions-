```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt 
import seaborn as sns   
data=pd.read_csv("/Users/princu_67/Downloads/market_data.csv")
#Q-1
print(data.head(10))  
print(data.shape)
print(data.info())
```

```python
data.isnull().sum()
```

```python
#Q-2
a=np.mean(data["Item_Weight"])
print(a)
data["Item_Weight"]=data["Item_Weight"].fillna(a)
data.head(10)

```

```python
#Q-3
b=data["Outlet_Size"].mode()
print(b)
data["Outlet_Size"]=data["Outlet_Size"].fillna("Medium")
data.head(10)
```

```python
#Q-4
data["Item_Fat_Content"].unique() 
a={"LF":"Low Fat", "low fat":"Low Fat", "reg":"Regular"}
data["Item_Fat_Content"]=data["Item_Fat_Content"].replace(a)
data["Item_Fat_Content"].unique()   
```

```python
#Q-5
data["Item_Visibility"].replace(0,data["Item_Visibility"].median())
```

```python
#Q-6
data["Outlet_Age"]=2024-data["Outlet_Establishment_Year"]
data["Outlet_Age"]
```

```python
#Q-7
df=pd.DataFrame(data, columns=["Item_Type","Item_MRP"])
df[((df["Item_Type"]=="Snack Foods") | (df["Item_Type"]=="Fruits and Vegetables")) & (df["Item_MRP"]>150)].head()

```

```python
#Q-8
data.groupby("Outlet_Type").agg({"Item_Outlet_Sales":"mean"})
```

```python
#Q-9
data.pivot_table(index="Outlet_Location_Type",columns=["Outlet_Size"],values="Item_Outlet_Sales")
```

```python
#Q-10
data.groupby("Item_Identifier").agg({"Item_Outlet_Sales":"max"}).sort_values(by="Item_Outlet_Sales",ascending=False).head()
```

```python
#Q-11
plt.hist(data["Item_MRP"],bins=30)
plt.xlabel("Maximum Retail Price")
plt.ylabel("Frequency")
plt.title("Item MRP")
plt.show()
```

```python
#Q-12
plt.scatter(x=data["Item_Visibility"],y=data["Item_Outlet_Sales"],c="red",alpha=0.5)
plt.grid()
plt.show()
```

```python
#Q-13
plt.bar(data=data, x="Outlet_Identifier",height="Item_Outlet_Sales")
plt.xticks(rotation=90)
plt.show()
```
